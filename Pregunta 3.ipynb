```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.integrate import solve_ivp
from scipy.fft import fft
```

```python
#Definimos el sistema
def Reaccion_quimica(t, z, l1, l2, u1, u2, k1, k2, D, t1, t2):
    x1, x2, y1, y2 = z
    return [-(x1**3-l1*x1-u1)-k1*y1-D*(x1-x2), -(x2**3-l2*x2-u2)-k2*y2-D*(x2-x1), (x1-y1)/t1-D*(y1-y2), (x2-y2)/t2-D*(y2-y1) ]
```

```python
#parametros fijos
u1=3
u2=3
l1=0.02
l2=2
k1=3.3
k2=10
t1=1
t2=1
```

```python
#Intervalo de tiempo
t_0=0
t_max=50
t = np.linspace(t_0, t_max, 1000)
```

```python
#Valor de D=0
D=0
```

```python
#Soluci√≥n problema con D=0
sol1 = solve_ivp(Reaccion_quimica, [t_0, t_max], [0, 0, 0, 0], args=(l1,l2,u1,u2,k1,k2,D,t1,t2),
                dense_output=True)
```

```python
#Grafico D=0
z1 = sol1.sol(t)
plt.plot(t, z1.T)
plt.xlabel('t')
plt.legend(['x1', 'x2','y1','y2'], shadow=True)
plt.title('Sistema Reacci√≥n quimica')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(z1[0],z1[2])
plt.title('Trayectoria de la soluci√≥n para D=0')
plt.xlabel('x1')
plt.ylabel('y_1')
plt.show()
```

```python
#Valor de D=0.22
D=0.22
sol2 = solve_ivp(Reaccion_quimica, [t_0, t_max], [0, 0, 0, 0], args=(l1,l2,u1,u2,k1,k2,D,t1,t2),
                dense_output=True)
z2 = sol2.sol(t)
plt.plot(t, z2.T)
plt.xlabel('t')
plt.legend(['x1', 'x2','y1','y2'], shadow=True)
plt.title('Sistema Reacci√≥n quimica')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(z2[0],z2[2])
plt.title('Trayectoria de la soluci√≥n para D=0.22')
plt.xlabel('x1')
plt.ylabel('y_1')
plt.show()
```

```python
#Valor de D=0.1
D=0.1
sol3 = solve_ivp(Reaccion_quimica, [t_0, t_max], [0, 0, 0, 0], args=(l1,l2,u1,u2,k1,k2,D,t1,t2),
                dense_output=True)
z3 = sol3.sol(t)
plt.plot(t, z3.T)
plt.xlabel('t')
plt.legend(['x1', 'x2','y1','y2'], shadow=True)
plt.title('Sistema Reacci√≥n quimica')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.integrate import solve_ivp
```

```python
#epsilon=0.0001
e=0.0001
w = 3
y, x = np.mgrid[-w:w:100j, -w:w:100j]
U = y
F = np.piecewise(y, [y<-1, (-1<=y)*(y<=1), y>1], [lambda y: -2-y, lambda y: y, lambda y: 2-y])
V = (F-x)/e

sns.set_style('whitegrid')
fig = plt.figure(figsize=(8, 8))
plt.streamplot(x, y, U, V, color=U)
plt.title('Diagrama de fase para epsilon=0.0001')
plt.show()
```

```python
#Definimos el sistema
def sistema(t, z, e):
    x, y = z
    F = np.piecewise(y, [y<-1, (-1<=y)and(y<=1), y>1], [lambda y: -2-y, lambda y: y, lambda y: 2-y])
    return [y, (F-x)/e]
```

```python
#Grafico y vs t
t0 = 0
tmax = 20
t = np.linspace(t0, tmax, 300)
plt.figure(figsize=(15,8))
for e in [0.0001,0.001,0.01,0.1]:
    sol = solve_ivp(sistema, [t0, tmax], [0.5, 0.5], args=(e,), dense_output=True)
    z = sol.sol(t)
    y = z[1]
    plt.plot(t, y, label=f'$\epsilon={e}$')

plt.xlabel('t')
plt.ylabel('y')
#plt.ylim(-3.2,4)
plt.legend(shadow=True)
plt.title('Variación de y vs t para los distintos valores de epsilon')
plt.show()
```
